'Moving Average - MA'
A moving average (MA) is a widely used indicator in technical analysis that helps smooth out price action by filtering out the “noise” from random price fluctuations. It is a trend-following, or lagging, indicator because it is based on past prices.


```python
## Lets go ahead and plot out several moving averages
ma_day = [10,20,50]

for ma in ma_day:
    col_name = "MA for %s days" %(str(ma))
    df[col_name] = pd.DataFrame.rolling(df['close'],ma).mean()
```

```python
df.head()
```

# Step 5: Feature Engineering

```python
df['MA20'] = df['close'].rolling(window=20).mean()
df['MA50'] = df['close'].rolling(window=50).mean()
df['Lag1'] = df['close'].shift(1)
df['Lag2'] = df['close'].shift(2)
df.dropna(inplace=True)
```

# Split the Data:

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
features = df[['MA20', 'MA50', 'Lag1', 'Lag2']]
target = df['close']
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

# Train Linear Regression Model:

```python
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
lr_predictions = lr_model.predict(X_test)
```

# Train Decision Tree Model:

```python
dt_model = DecisionTreeRegressor(random_state=42)
dt_model.fit(X_train, y_train)
dt_predictions = dt_model.predict(X_test)
```

# Train Random Forest Model:

```python
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
rf_predictions = rf_model.predict(X_test)
```

# Evaluate Models:

```python
def evaluate_model(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    return mae, mse, rmse

lr_mae, lr_mse, lr_rmse = evaluate_model(y_test, lr_predictions)
dt_mae, dt_mse, dt_rmse = evaluate_model(y_test, dt_predictions)
rf_mae, rf_mse, rf_rmse = evaluate_model(y_test, rf_predictions)

print(f'Linear Regression - MAE: {lr_mae}, MSE: {lr_mse}, RMSE: {lr_rmse}')
print(f'Decision Tree - MAE: {dt_mae}, MSE: {dt_mse}, RMSE: {dt_rmse}')
print(f'Random Forest - MAE: {rf_mae}, MSE: {rf_mse}, RMSE: {rf_rmse}')
```

# Select the Best Model:

```python
model_errors = {
    'Linear Regression': lr_rmse,
    'Decision Tree': dt_rmse,
    'Random Forest': rf_rmse
}
 
best_model_name = min(model_errors, key=model_errors.get)
best_model_rmse = model_errors[best_model_name]

print(f'The best model is {best_model_name} with RMSE: {best_model_rmse}')
```

# Visualize Predictions of the Best Model:

```python
best_predictions = rf_predictions if best_model_name == 'Random Forest' else dt_predictions if best_model_name == 'Decision Tree' else lr_predictions

plt.figure(figsize=(14, 7))
plt.plot(y_test.index, y_test, label='Actual')
plt.plot(y_test.index, best_predictions, label='Predicted')
plt.title(f'Stock Price Prediction - {best_model_name}')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.show()
```
